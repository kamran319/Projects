```python
# legit and fraud trsanction, an example of unbalanced data, 0 represent legit and 1 represent fraudulent
cc["Class"].value_counts()
```

```python
legit=cc[cc.Class==0]
fraud=cc[cc.Class==1]
```

```python
# Under sampling
# Build a sample data like similar distributions of noraml trasaction and fraudulent
# Number of fradulent trasactions--> 492
legit_sample=legit.sample(n=492)
```

```python
# Concatenating two Dataframe
new_data=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_data.head()
```

```python
new_data.tail()
```

```python
new_data['Class'].value_counts()
```

```python
new_data.groupby('Class').mean()
```

```python
X=new_data.drop(['Class'],axis=1)
y=new_data['Class']
```

```python
X.head()
```

```python
y.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y,random_state=2)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X.shape
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
training_data_prediction=model.predict(X_train)
training_data_acc=accuracy_score(y_train,training_data_prediction)
print(training_data_acc)
```

```python
testing_data_prediction=model.predict(X_test)
testing_data_acc=accuracy_score(y_test,testing_data_prediction)
print(testing_data_acc)
```
